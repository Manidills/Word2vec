```
import numpy as np
import re
import gensim
```

```
with open('/content/Extractive-based-summarization/rahul dravid.txt','r') as dravid:
    lines=list(dravid)

    
```

```
print(lines)
```

```
def preprocess(text):
    new=text.lower()
    new=re.sub("[\(\[].*?[\)\]]", "", new)
    new=re.sub("'s","",new)
    new=re.sub("[^'0-9.a-zA-Z]", " ", new)
    token=new.split()
    return (" ".join(token)).strip()
```

```
word=[]

for i in lines:
    word.append(preprocess(i))
```

```
word
```

```
sent=[]
for i in word:
    sentence=i.split(".")
    for i in sentence:
        if(i!=''):
            sent.append(i.strip())
       
```

```
word_list = nltk.word_tokenize(sentence)
```

```
word_list
```

```
from nltk.stem import WordNetLemmatizer 
```

```
lemmatizer = WordNetLemmatizer()
```

```
lemmatized_output = ' '.join([lemmatizer.lemmatize(w) for w in word_list])
```

```
lemmatized_output
```

```
model_w2v = gensim.models.Word2Vec(
            lemmatized_output,
            size=200, # desired no. of features/independent variables 
            window=5, # context window size
            min_count=2,
            sg = 0, # 1 for cbow model
            hs = 0,
            negative = 10, # for negative sampling
            workers= 2, # no.of cores
seed = 34)
```

```
model_w2v.train(lemmatized_output, total_examples= len(lemmatized_output), epochs=model_w2v.epochs)
```
